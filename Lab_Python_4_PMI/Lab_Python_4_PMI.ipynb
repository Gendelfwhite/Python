*apple – лесть, яблоко  
soap – лесть, мыло

```python
input_file_path = '/content/drive/MyDrive/Программирование на языке Python/Lab_Python_4_PMI/input.txt'
output_file_path = 'output.txt'

ru_en_dict = {}
bad_lines = []

def is_valid_line(line):
    line = line.strip()
    if not line:
        return False, "Пустая строка"

    # Нормализуем все тире/дефисы к одному виду для проверки
    normalized_line = line.replace(' - ', ' – ').replace(' — ', ' – ')

    # Проверяем количество нормализованных разделителей
    if normalized_line.count(' – ') != 1:
        return False, "Некорректное количество разделителей (должен быть ровно один ' – ' или его аналоги)"

    # Разделяем по первому найденному разделителю (любого типа)
    separator_pos = max(line.find(' – '), line.find(' - '), line.find(' — '))
    if separator_pos == -1:
        return False, "Отсутствует корректный разделитель"

    parts = [line[:separator_pos], line[separator_pos+3:]]
    en_word = parts[0].strip()
    ru_words_part = parts[1].strip()

    if not en_word:
        return False, "Отсутствует английское слово"
    if not ru_words_part:
        return False, "Отсутствуют русские переводы"

    return True, ""

try:
    with open(input_file_path, 'r', encoding='utf-8') as file:
        for line_num, line in enumerate(file, 1):
            valid, error = is_valid_line(line)

            if not valid:
                bad_lines.append((line_num, line.strip(), error))
                continue

            # Находим позицию первого разделителя любого типа
            separator_pos = max(line.find(' – '), line.find(' - '), line.find(' — '))
            en_word = line[:separator_pos].strip()
            ru_words_part = line[separator_pos+3:].strip()

            ru_words = [word.strip() for word in ru_words_part.split(',') if word.strip()]

            for ru_word in ru_words:
                if ru_word not in ru_en_dict:
                    ru_en_dict[ru_word] = []
                if en_word not in ru_en_dict[ru_word]:
                    ru_en_dict[ru_word].append(en_word)

    # Запись результатов в файл
    with open(output_file_path, 'w', encoding='utf-8') as file:
        for ru_word in sorted(ru_en_dict.keys()):
            en_words = sorted(ru_en_dict[ru_word])
            file.write(f"{ru_word} – {', '.join(en_words)}\n")

    # Вывод информации о плохих строках
    if bad_lines:
        print("\nОбнаружены некорректные строки:")
        for line_num, line, error in bad_lines:
            print(f"Строка {line_num}: '{line}' - Ошибка: {error}")
    else:
        print("Все строки обработаны успешно.")

    print(f"\nРусско-английский словарь сохранён в {output_file_path}")

except FileNotFoundError:
    print(f"Ошибка: Файл не найден: {input_file_path}")
except Exception as e:
    print(f"Произошла непредвиденная ошибка: {str(e)}")
```

```python
from google.colab import drive
import os
import random

try:
    # Подключение Google Диска
    drive.mount('/content/drive')
except Exception as e:
    print(f"Ошибка при подключении Google Диска: {e}")
    exit()

# Путь к папке с файлами анкет
directory = '/content/drive/MyDrive/Программирование на языке Python/Lab_Python_4_PMI/friends_catalog'
friends_catalog = {}

# Чтение данных из всех файлов в указанной директории
try:
    if not os.path.exists(directory):
        raise FileNotFoundError(f"Папка не найдена: {directory}")

    for filename in os.listdir(directory):
        if filename.endswith(".txt"):
            file_path = os.path.join(directory, filename)
            try:
                with open(file_path, 'r', encoding='utf-8') as file:
                    lines = file.readlines()
                    if len(lines) < 17:
                        print(f"Файл {filename} пропущен: недостаточно строк.")
                        continue

                    # Извлечение данных по индексам
                    name = lines[1].strip()
                    surname = lines[4].strip()
                    full_name = f"{name} {surname}"

                    age = lines[7].strip()
                    favorite_food = lines[10].strip()
                    music_band = lines[13].strip()
                    dream = lines[16].strip()

                    friends_catalog[full_name] = {
                        "Возраст": age,
                        "Любимая еда": favorite_food,
                        "Музыкальная группа": music_band,
                        "Заветная мечта": dream
                    }
            except Exception as file_error:
                print(f"Ошибка при чтении файла {filename}: {file_error}")

    if not friends_catalog:
        print("Каталог друзей пуст. Проверьте содержимое папки.")
        exit()

except Exception as dir_error:
    print(f"Ошибка при обработке директории: {dir_error}")
    exit()

print("Добро пожаловать в игру 'Знаешь ли ты своих друзей?'!")

# Основной игровой цикл
while True:
    try:
        # Проверка наличия друзей
        if not friends_catalog:
            print("Нет доступных анкет для игры.")
            break

        # Случайный выбор друга
        friend_name, friend_info = random.choice(list(friends_catalog.items()))

        # Получаем и перемешиваем категории
        categories = list(friend_info.keys())
        random.shuffle(categories)

        print("\nПопробуйте угадать, кто это!")
        guessed = False

        for category in categories:
            value = friend_info[category]
            print(f"\nКатегория: {category}")
            print(f"Подсказка: {value}")

            answer = input("Кто это? (Введите имя и фамилию): ").strip()
            if answer.lower() == friend_name.lower():
                print("Правильно! Вы угадали!")
                guessed = True
                break
            else:
                print("Неверно. Попробуйте еще раз.")

        if not guessed:
            print(f"\nВы не смогли угадать. Это был(а) {friend_name}.")

        # Повтор игры
        choice = input("\nХотите попробовать снова? (да/нет): ").strip().lower()
        if choice != 'да':
            print("Спасибо за игру! До свидания!")
            break

    except KeyboardInterrupt:
        print("\nИгра прервана пользователем.")
        break
    except Exception as e:
        print(f"Произошла ошибка в игровом цикле: {e}")
        break
```

**3.**	Создайте словарь, где ключами являются числа, а значениями – строки. Примените к нему метод items(), полученный объект dict_items передайте в написанную вами функцию, которая создает и возвращает новый словарь, обратный исходному, т.е. ключами являются строки, а значениями – числа (если соответствует только одно значение) или списки чисел (если несколько).

```python
def reverse_dict(items):
    """
    Функция для создания обратного словаря.

    :param items: Объект dict_items из исходного словаря
    :return: Обратный словарь
    """
    reversed_dict = {}

    try:
        for key, value in items:
            try:
                # Попытка использовать значение как ключ
                if value not in reversed_dict:
                    reversed_dict[value] = [key]
                else:
                    reversed_dict[value].append(key)
            except TypeError as te:
                print(f"Ошибка: значение '{value}' не может быть использовано как ключ (нехешируемо). Пропущено.")
                continue

        # Преобразуем списки из одного элемента в простые значения
        for key in list(reversed_dict.keys()):
            if len(reversed_dict[key]) == 1:
                reversed_dict[key] = reversed_dict[key][0]

        return reversed_dict

    except TypeError as e:
        print(f"Ошибка: входные данные должны быть итерируемыми парами (key, value).")
        print(f"Подробности: {e}")
        return {}
    except Exception as e:
        print(f"Произошла непредвиденная ошибка при развороте словаря: {e}")
        return {}

# Создаем исходный словарь
original_dict = {
    1: "apple",
    2: "banana",
    3: "apple",
    4: "cherry",
    5: "banana",
    6: "apple"
}

# Получаем объект dict_items
try:
    items = original_dict.items()
except AttributeError as e:
    print(f"Ошибка: объект не поддерживает метод .items(): {e}")
    items = []

# Вызываем функцию для создания обратного словаря
reversed_dict = reverse_dict(items)

# Выводим результаты
print("Исходный словарь:")
print(original_dict)
print("\nОбратный словарь:")
print(reversed_dict)
```

**4.**	Имеется каталог X, содержащий файлы и другие каталоги. Создать копию структуры каталогов, где каждый файл *.txt обрабатывается следующим образом: первой строкой вставляется число строк в исходном файле, а последней строкой – уровень вложенности в каталог X. Остальные файлы пропускаются. Сам каталог X и его содержимое изменяться не должны.

```python
import os
import shutil

# Путь к исходному каталогу X на Google Диске
source_dir = '/content/drive/MyDrive/Программирование на языке Python/Lab_Python_4_PMI/X'

# Путь к копии структуры каталогов
target_dir = '/content/drive/MyDrive/Программирование на языке Python/Lab_Python_4_PMI/copy_of_X'

# Функция для определения уровня вложенности
def get_nesting_level(path, base_dir):
    try:
        relative_path = os.path.relpath(path, base_dir)
        return relative_path.count(os.sep)
    except Exception as e:
        print(f"Ошибка при определении уровня вложенности для {path}: {e}")
        return 0

# Функция для обработки .txt-файла
def process_txt_file(source_file, target_file):
    try:
        with open(source_file, 'r', encoding='utf-8') as f:
            lines = f.readlines()
            num_lines = len(lines)
    except Exception as e:
        print(f"Ошибка при чтении файла {source_file}: {e}")
        return

    nesting_level = get_nesting_level(source_file, source_dir)

    try:
        with open(target_file, 'w', encoding='utf-8') as f:
            f.write(f"{num_lines}\n")
            f.writelines(lines)
            f.write(f"{nesting_level}\n")
    except Exception as e:
        print(f"Ошибка при записи файла {target_file}: {e}")

# Функция для рекурсивного копирования и обработки файлов
def copy_and_process_directory(source, target):
    try:
        if not os.path.exists(target):
            os.makedirs(target)
    except Exception as e:
        print(f"Ошибка при создании каталога {target}: {e}")
        return

    try:
        for item in os.listdir(source):
            source_item = os.path.join(source, item)
            target_item = os.path.join(target, item)

            if os.path.isdir(source_item):
                copy_and_process_directory(source_item, target_item)
            elif source_item.endswith('.txt'):
                process_txt_file(source_item, target_item)
            else:
                try:
                    shutil.copy2(source_item, target_item)
                except Exception as e:
                    print(f"Ошибка при копировании файла {source_item}: {e}")
    except Exception as e:
        print(f"Ошибка при обработке содержимого каталога {source}: {e}")

# Главный блок выполнения
if __name__ == "__main__":
    try:
        if not os.path.exists(source_dir):
            raise FileNotFoundError(f"Исходный каталог не найден: {source_dir}")

        copy_and_process_directory(source_dir, target_dir)
        print(f"Копия структуры каталогов создана в {target_dir}")
    except Exception as e:
        print(f"Произошла ошибка при выполнении скрипта: {e}")
```

```python
import os
import shutil

# Путь к исходному каталогу X на Google Диске
source_dir = '/content/drive/MyDrive/Программирование на языке Python/Lab_Python_4_PMI/X'

# Путь к копии структуры каталогов
target_dir = '/content/drive/MyDrive/Программирование на языке Python/Lab_Python_4_PMI/copy_of_X'

def get_nesting_level(path, base_dir):
    """Возвращает уровень вложенности файла относительно базового каталога"""
    try:
        relative_path = os.path.relpath(os.path.dirname(path), base_dir)
        if relative_path == '.':
            return 0
        return relative_path.count(os.sep) + 1
    except Exception as e:
        print(f"Ошибка при определении уровня вложенности для {path}: {e}")
        return 0

def process_txt_file(source_file, target_file):
    """Обрабатывает текстовый файл, добавляя метаданные в начало"""
    try:
        with open(source_file, 'r', encoding='utf-8') as f:
            content = f.read()
            lines = content.splitlines()
            num_lines = len(lines)
            nesting_level = get_nesting_level(source_file, source_dir)

        with open(target_file, 'w', encoding='utf-8') as f:
            f.write(f"Число строк в исходном файле: {num_lines}\n")
            f.write(f"Уровень вложенности в каталог X: {nesting_level}\n")
            f.write(content)

    except Exception as e:
        print(f"Ошибка при обработке файла {source_file}: {e}")

def copy_and_process_directory(source, target):
    """Рекурсивно копирует структуру каталогов"""
    try:
        os.makedirs(target, exist_ok=True)
    except Exception as e:
        print(f"Ошибка при создании каталога {target}: {e}")
        return

    try:
        for item in os.listdir(source):
            source_item = os.path.join(source, item)
            target_item = os.path.join(target, item)

            if os.path.isdir(source_item):
                copy_and_process_directory(source_item, target_item)
            elif item.lower().endswith('.txt'):
                process_txt_file(source_item, target_item)
            else:
                try:
                    shutil.copy2(source_item, target_item)
                except Exception as e:
                    print(f"Ошибка при копировании файла {source_item}: {e}")
    except Exception as e:
        print(f"Ошибка при обработке содержимого каталога {source}: {e}")

if __name__ == "__main__":
    try:
        if not os.path.exists(source_dir):
            raise FileNotFoundError(f"Исходный каталог не найден: {source_dir}")

        print("Начало обработки...")
        copy_and_process_directory(source_dir, target_dir)
        print(f"Копия структуры каталогов успешно создана в {target_dir}")
    except Exception as e:
        print(f"Произошла критическая ошибка: {e}")
```

**5.** Напишите программу, которая для заданного каталога создает подкаталоги, соответствующие дате создания каждого отдельного файла, и перемещает каждый файл в соответствующий дате каталог.

```python
import os
import shutil
from datetime import datetime

def organize_files(source_dir):
    """
    Перемещает файлы из указанной директории в подпапки по дате создания.
    """
    try:
        # Проверка существования директории
        if not os.path.exists(source_dir):
            raise FileNotFoundError(f"Каталог не найден: {source_dir}")

        # Получаем список всех объектов в папке
        files = os.listdir(source_dir)

        if not files:
            print("Каталог пуст, нечего сортировать.")
            return

        # Обрабатываем каждый файл
        for file in files:
            file_path = os.path.join(source_dir, file)

            # Пропускаем папки, работаем только с файлами
            if not os.path.isfile(file_path):
                continue

            try:
                # Получаем время создания и преобразуем в дату
                creation_time = os.path.getctime(file_path)
                date = datetime.fromtimestamp(creation_time).strftime('%Y-%m-%d')

                # Папка с датой
                date_folder = os.path.join(source_dir, date)

                # Создаем папку, если не существует
                if not os.path.exists(date_folder):
                    os.makedirs(date_folder)

                # Перемещаем файл
                new_path = os.path.join(date_folder, file)
                shutil.move(file_path, new_path)
                print(f"Файл '{file}' перемещён в папку '{date}'")

            except PermissionError:
                print(f"Ошибка доступа к файлу: {file}. Пропущен.")
            except Exception as e:
                print(f"Ошибка при обработке файла '{file}': {e}")

    except FileNotFoundError as fnf_error:
        print(fnf_error)
    except Exception as e:
        print(f"Произошла неожиданная ошибка: {e}")

# Укажите путь к папке
folder_path = '/content/drive/MyDrive/Программирование на языке Python/Lab_Python_4_PMI/5_задание'

# Запускаем программу
print("Начинаем сортировку файлов...")
organize_files(folder_path)
print("Сортировка завершена!")
```

```python
import os
import shutil
from datetime import datetime
import pytz

def organize_files_by_time(source_dir):
    """
    Организует файлы в подкаталоги по времени создания (только в коллабе)
    """
    try:
        if not os.path.exists(source_dir):
            raise FileNotFoundError(f"Каталог не найден: {source_dir}")

        items = os.listdir(source_dir)
        if not items:
            print("Каталог пуст, нечего сортировать.")
            return

        processed_files = 0
        skipped_files = 0

        # Устанавливаем временную зону (для коллаба)
        timezone = pytz.timezone('Europe/Moscow')

        for item in items:
            item_path = os.path.join(source_dir, item)

            if not os.path.isfile(item_path):
                continue

            try:
                # Получаем метаданные файла
                stat = os.stat(item_path)

                # Для коллаба используем st_mtime как время создания
                creation_timestamp = stat.st_mtime

                # Конвертируем с учетом временной зоны
                time_obj = datetime.fromtimestamp(creation_timestamp, timezone)

                # Форматируем время в ЧЧ_ММ (24-часовой формат)
                time_folder = time_obj.strftime("%H_%M")
                actual_time = time_obj.strftime("%H:%M")

                # Проверка корректности времени
                if not (0 <= time_obj.hour < 24 and 0 <= time_obj.minute < 60):
                    raise ValueError(f"Некорректное время файла: {actual_time}")

                # Создаем целевую папку
                target_dir = os.path.join(source_dir, time_folder)
                os.makedirs(target_dir, exist_ok=True)

                # Перемещаем файл
                new_path = os.path.join(target_dir, item)
                shutil.move(item_path, new_path)
                print(f"Файл '{item}' перемещён в '{time_folder}/' (фактическое время: {actual_time})")
                processed_files += 1

            except PermissionError:
                print(f"Ошибка доступа: {item} (пропущен)")
                skipped_files += 1
            except Exception as e:
                print(f"Ошибка обработки '{item}': {str(e)}")
                skipped_files += 1

        print(f"\nСтатистика:")
        print(f"Обработано файлов: {processed_files}")
        print(f"Пропущено файлов: {skipped_files}")

    except FileNotFoundError as fnf_error:
        print(fnf_error)
    except Exception as e:
        print(f"Критическая ошибка: {str(e)}")

# Укажите путь для сортировки
folder_to_organize = '/content/drive/MyDrive/Программирование на языке Python/Lab_Python_4_PMI/5_задание'

# Установка временной зоны для Colab (если нужно)
!pip install pytz
import pytz

print("Начинаем сортировку файлов по времени создания...")
organize_files_by_time(folder_to_organize)
print("Процесс завершен! Проверьте результат.")
```

```python
import os
import shutil
from datetime import datetime
import pytz

def ask_user_action(file_path):
    """
    Запрашивает действие у пользователя при конфликте имен файлов
    """
    while True:
        print(f"\nФайл '{os.path.basename(file_path)}' уже существует.")
        print("1. Перезаписать")
        print("2. Пропустить")
        print("3. Переименовать")
        choice = input("Выберите действие (1-3): ").strip()

        if choice == '1':
            return 'overwrite'
        elif choice == '2':
            return 'skip'
        elif choice == '3':
            return 'rename'
        else:
            print("Некорректный ввод. Попробуйте снова.")

def organize_files_by_time(source_dir):
    """
    Организует файлы в подкаталоги по времени создания с проверкой конфликтов
    """
    try:
        if not os.path.exists(source_dir):
            raise FileNotFoundError(f"Каталог не найден: {source_dir}")

        items = os.listdir(source_dir)
        if not items:
            print("Каталог пуст, нечего сортировать.")
            return

        processed_files = 0
        skipped_files = 0
        renamed_files = 0

        timezone = pytz.timezone('Europe/Moscow')

        for item in items:
            item_path = os.path.join(source_dir, item)

            if not os.path.isfile(item_path):
                continue

            try:
                stat = os.stat(item_path)
                creation_timestamp = stat.st_mtime
                time_obj = datetime.fromtimestamp(creation_timestamp, timezone)
                time_folder = time_obj.strftime("%H_%M")
                actual_time = time_obj.strftime("%H:%M")

                if not (0 <= time_obj.hour < 24 and 0 <= time_obj.minute < 60):
                    raise ValueError(f"Некорректное время файла: {actual_time}")

                target_dir = os.path.join(source_dir, time_folder)
                os.makedirs(target_dir, exist_ok=True)

                new_path = os.path.join(target_dir, item)

                # Обработка случая, когда файл уже существует
                if os.path.exists(new_path):
                    action = ask_user_action(new_path)

                    if action == 'skip':
                        print(f"Файл '{item}' пропущен.")
                        skipped_files += 1
                        continue
                    elif action == 'rename':
                        base, ext = os.path.splitext(item)
                        counter = 1
                        while os.path.exists(new_path):
                            new_name = f"{base}_{counter}{ext}"
                            new_path = os.path.join(target_dir, new_name)
                            counter += 1
                        renamed_files += 1
                        print(f"Файл переименован в '{os.path.basename(new_path)}'")

                shutil.move(item_path, new_path)
                print(f"Файл '{os.path.basename(new_path)}' перемещён в '{time_folder}/' (время создания: {actual_time})")
                processed_files += 1

            except PermissionError:
                print(f"Ошибка доступа: {item} (пропущен)")
                skipped_files += 1
            except Exception as e:
                print(f"Ошибка обработки '{item}': {str(e)}")
                skipped_files += 1

        print(f"\nСтатистика:")
        print(f"Обработано файлов: {processed_files}")
        print(f"Переименовано файлов: {renamed_files}")
        print(f"Пропущено файлов: {skipped_files}")

    except FileNotFoundError as fnf_error:
        print(fnf_error)
    except Exception as e:
        print(f"Критическая ошибка: {str(e)}")

# Установка зависимостей (если нужно)
try:
    import pytz
except ImportError:
    !pip install pytz
    import pytz

# Укажите путь для сортировки
folder_to_organize = '/content/drive/MyDrive/Программирование на языке Python/Lab_Python_4_PMI/5_задание'

print("\nНачинаем сортировку файлов по времени создания...")
organize_files_by_time(folder_to_organize)
print("\nПроцесс завершен! Проверьте результат.")
```

**6.**	Из файлов *shop1.txt* и *shop2.txt* считываются прайс-листы (наименование товара – цена). Необходимо объединить два прайс-листа так, чтобы если наименование товара присутствует в обоих листах, то в итоговый прайс-лист помещается большая цена. Результат записать в файл *shop_max.txt*.

```python
import re
import os

# Пути к файлам
shop1_path = "/content/drive/MyDrive/Программирование на языке Python/Lab_Python_4_PMI/6_задание/shop1.txt"
shop2_path = "/content/drive/MyDrive/Программирование на языке Python/Lab_Python_4_PMI/6_задание/shop2.txt"
output_path = "/content/drive/MyDrive/Программирование на языке Python/Lab_Python_4_PMI/6_задание/shop_max.txt"


def parse_price_list(file_path):
    """
    Функция для извлечения товаров и цен из файла.
    Возвращает словарь вида {товар: цена}.
    """
    price_dict = {}

    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            for line_num, line in enumerate(file, start=1):
                line = re.sub(r'^\d+\.\s+', '', line)  # Убираем номер строки

                # Ищем название товара и цену
                match = re.search(r'(.+?)\s+\.+\s+(\d+)\s+руб\.', line)
                if match:
                    item = match.group(1).strip()
                    price = int(match.group(2))
                    price_dict[item] = price
                else:
                    print(f"Предупреждение: строка {line_num} в файле '{file_path}' не соответствует формату и будет пропущена.")
    except FileNotFoundError:
        print(f"Ошибка: файл '{file_path}' не найден.")
    except PermissionError:
        print(f"Ошибка: нет доступа к файлу '{file_path}'.")
    except Exception as e:
        print(f"Произошла ошибка при чтении файла '{file_path}': {e}")

    return price_dict


def write_price_list_to_file(price_dict, output_path):
    """
    Записывает объединённый прайс-лист в файл.
    """
    try:
        max_item_length = max((len(item) for item in price_dict), default=0)

        with open(output_path, 'w', encoding='utf-8') as file:
            file.write("======================================\n")
            file.write("      ОБЪЕДИНЕННЫЙ ПРАЙС-ЛИСТ\n")
            file.write("======================================\n\n")

            for i, (item, price) in enumerate(sorted(price_dict.items()), start=1):
                dots = '.' * (max_item_length - len(item) + 10)
                file.write(f"{i}. {item} {dots} {price} руб.\n")

        print(f"Файл '{output_path}' успешно создан!")

    except PermissionError:
        print(f"Ошибка: нет прав на запись в файл '{output_path}'.")
    except Exception as e:
        print(f"Произошла ошибка при записи в файл '{output_path}': {e}")


# Главный блок
def main():
    print("Чтение прайс-листов...")

    shop1_prices = parse_price_list(shop1_path)
    shop2_prices = parse_price_list(shop2_path)

    if not shop1_prices and not shop2_prices:
        print("Ошибка: Оба прайс-листа пустые или содержат ошибки форматирования.")
        return

    # Объединяем прайс-листы с выбором максимальной цены
    merged_prices = shop1_prices.copy()
    for item, price in shop2_prices.items():
        merged_prices[item] = max(merged_prices.get(item, 0), price)

    # Записываем в файл
    write_price_list_to_file(merged_prices, output_path)


# Запуск программы
if __name__ == "__main__":
    main()
```

**7.** Создайте файл *input.txt*, содержащий данные о странах экспортерах (не менее 10): название страны, название экспортируемого товара и объем экспортируемого товара (в тоннах). Напишите программу, позволяющую обрабатывать следующие запросы пользователя:
- добавить данные в файл;

- вывести всю информацию, содержащуюся в файле;

- вывести список товаров, экспортируемых введенной пользователем страной;

- вывести список стран, экспортирующих заданный товар.

Если добавляется информация о уже имеющихся в базе стране и товаре, новое значение тонн должно прибавлять к уже имеющемуся количеству.

```python
file_path = "/content/drive/MyDrive/Программирование на языке Python/Lab_Python_4_PMI/7_задание/input.txt"

def read_data():
    """Функция для чтения данных из файла"""
    data = {}
    try:
        with open(file_path, "r", encoding="utf-8") as file:
            for line in file:
                parts = line.strip().split("; ")
                if len(parts) != 3:
                    raise ValueError(f"Неверный формат строки: {line}")
                country, product, quantity_str = parts
                quantity = int(quantity_str)
                if country not in data:
                    data[country] = {}
                if product in data[country]:
                    data[country][product] += quantity  # Суммируем тонны
                else:
                    data[country][product] = quantity
    except FileNotFoundError:
        print(f"Ошибка: файл {file_path} не найден.")
    except ValueError as e:
        print(f"Ошибка: {e}")
    except Exception as e:
        print(f"Произошла непредвиденная ошибка при чтении файла: {e}")
    return data

# Функция для записи данных в файл
def write_data(data):
    """Функция для записи данных в файл"""
    try:
        with open(file_path, "w", encoding="utf-8") as file:
            for country, products in data.items():
                for product, quantity in products.items():
                    file.write(f"{country}; {product}; {quantity}\n")
    except IOError:
        print(f"Ошибка ввода-вывода при записи в файл {file_path}.")
    except Exception as e:
        print(f"Произошла непредвиденная ошибка при записи в файл: {e}")

# Функция для вывода всех данных
def show_all():
    """Функция для вывода всех данных"""
    data = read_data()
    print("\n ВЕСЬ СПИСОК ЭКСПОРТА:")
    for country, products in data.items():
        for product, quantity in products.items():
            print(f"{country}: {product} - {quantity} тонн")

# Функция для добавления данных
def add_data():
    """Функция для добавления данных"""
    try:
        country = input("Введите страну: ").strip()
        product = input("Введите экспортируемый товар: ").strip()
        quantity = int(input("Введите объем экспорта (тонн): "))

        data = read_data()

        # Если страна уже есть, обновляем данные
        if country in data:
            if product in data[country]:
                data[country][product] += quantity
            else:
                data[country][product] = quantity
        else:
            data[country] = {product: quantity}

        write_data(data)
        print(f" Добавлено: {country} - {product} ({quantity} тонн)")
    except ValueError:
        print("Ошибка: неверный формат ввода. Объем экспорта должен быть числом.")
    except Exception as e:
        print(f"Произошла непредвиденная ошибка при добавлении данных: {e}")

# Функция для поиска товаров по стране
def get_products_by_country():
    """Функция для поиска товаров по стране"""
    try:
        country = input("Введите название страны: ").strip()
        data = read_data()

        if country in data:
            print(f"\n {country} экспортирует:")
            for product, quantity in data[country].items():
                print(f"- {product}: {quantity} тонн")
        else:
            print(" Страна не найдена.")
    except Exception as e:
        print(f"Произошла непредвиденная ошибка при поиске товаров по стране: {e}")

# Функция для поиска стран по товару
def get_countries_by_product():
    """Функция для поиска стран по товару"""
    try:
        product = input("Введите название товара: ").strip()
        data = read_data()
        countries = []

        for country, products in data.items():
            if product in products:
                countries.append(f"{country} ({products[product]} тонн)")

        if countries:
            print(f"\n {product} экспортируют:")
            for country in countries:
                print(f"- {country}")
        else:
            print(" Товар не найден.")
    except Exception as e:
        print(f"Произошла непредвиденная ошибка при поиске стран по товару: {e}")

def menu():
    """Функция для отображения меню и обработки выбора пользователя"""
    while True:
        print("\n Меню:")
        print("1. Вывести всю информацию")
        print("2. Добавить данные")
        print("3. Найти товары по стране")
        print("4. Найти страны по товару")
        print("5. Выход")

        choice = input("Выберите действие (1-5): ")

        if choice == "1":
            show_all()
        elif choice == "2":
            add_data()
        elif choice == "3":
            get_products_by_country()
        elif choice == "4":
            get_countries_by_product()
        elif choice == "5":
            print("Выход из программы.")
            break
        else:
            print("Неверный ввод! Попробуйте снова.")

# Запуск меню
menu()
```

```python
file_path = "/content/drive/MyDrive/Программирование на языке Python/Lab_Python_4_PMI/7_задание/input1.txt"

def is_valid_country(country):
    """Проверка корректности названия страны"""
    if not country or country.isspace():
        return False, "Название страны не может быть пустым"
    if country.isdigit():
        return False, "Название страны не может состоять только из цифр"
    return True, ""

def is_valid_product(product):
    """Проверка корректности названия товара"""
    if not product or product.isspace():
        return False, "Название товара не может быть пустым"
    if product.isdigit():
        return False, "Название товара не может состоять только из цифр"
    return True, ""

def is_valid_quantity(quantity_str):
    """Проверка корректности количества"""
    if not quantity_str or quantity_str.isspace():
        return False, "Количество не может быть пустым"
    if not quantity_str.isdigit():
        return False, "Количество должно быть целым числом"
    if int(quantity_str) <= 0:
        return False, "Количество должно быть положительным числом"
    return True, ""

def read_data():
    """Функция для чтения данных из файла с проверками"""
    data = {}
    bad_lines = []

    try:
        with open(file_path, "r", encoding="utf-8") as file:
            for line_num, line in enumerate(file, 1):
                line = line.strip()
                if not line:
                    continue

                parts = line.split("; ")
                if len(parts) != 3:
                    bad_lines.append((line_num, line, "Неверный формат строки (должно быть 3 части)"))
                    continue

                country, product, quantity_str = parts
                errors = []

                # Проверка страны
                valid, msg = is_valid_country(country)
                if not valid:
                    errors.append(msg)

                # Проверка товара
                valid, msg = is_valid_product(product)
                if not valid:
                    errors.append(msg)

                # Проверка количества
                valid, msg = is_valid_quantity(quantity_str)
                if not valid:
                    errors.append(msg)

                if errors:
                    bad_lines.append((line_num, line, " | ".join(errors)))
                else:
                    quantity = int(quantity_str)
                    if country not in data:
                        data[country] = {}
                    data[country][product] = data[country].get(product, 0) + quantity

    except FileNotFoundError:
        print(f"Ошибка: файл {file_path} не найден. Создан новый словарь данных.")
    except Exception as e:
        print(f"Произошла непредвиденная ошибка при чтении файла: {e}")

    # Вывод информации о некорректных строках
    if bad_lines:
        print("\nОбнаружены некорректные строки:")
        for line_num, line, error in bad_lines:
            print(f"Строка {line_num}: '{line}' - Ошибка: {error}")

    return data

def write_data(data):
    """Функция для записи данных в файл"""
    try:
        with open(file_path, "w", encoding="utf-8") as file:
            for country, products in data.items():
                for product, quantity in products.items():
                    file.write(f"{country}; {product}; {quantity}\n")
    except IOError:
        print(f"Ошибка ввода-вывода при записи в файл {file_path}.")
    except Exception as e:
        print(f"Произошла непредвиденная ошибка при записи в файл: {e}")

def show_all():
    """Функция для вывода всех данных"""
    data = read_data()
    if not data:
        print("Нет корректных данных для отображения.")
        return

    print("\nКОРРЕКТНЫЕ ДАННЫЕ ЭКСПОРТА:")
    for country, products in sorted(data.items()):
        print(f"\n{country}:")
        for product, quantity in sorted(products.items()):
            print(f"  {product}: {quantity} тонн")

def add_data():
    """Функция для добавления данных с проверками"""
    try:
        # Ввод и проверка страны
        while True:
            country = input("Введите страну: ").strip()
            valid, msg = is_valid_country(country)
            if valid:
                break
            print(f"Ошибка: {msg}")

        # Ввод и проверка товара
        while True:
            product = input("Введите экспортируемый товар: ").strip()
            valid, msg = is_valid_product(product)
            if valid:
                break
            print(f"Ошибка: {msg}")

        # Ввод и проверка количества
        while True:
            quantity_str = input("Введите объем экспорта (тонн): ").strip()
            valid, msg = is_valid_quantity(quantity_str)
            if valid:
                quantity = int(quantity_str)
                break
            print(f"Ошибка: {msg}")

        data = read_data()

        if country in data:
            if product in data[country]:
                data[country][product] += quantity
                print(f"Обновлено: {country} - {product} (+{quantity} тонн)")
            else:
                data[country][product] = quantity
                print(f"Добавлен новый товар: {country} - {product} ({quantity} тонн)")
        else:
            data[country] = {product: quantity}
            print(f"Добавлена новая запись: {country} - {product} ({quantity} тонн)")

        write_data(data)

    except Exception as e:
        print(f"Произошла непредвиденная ошибка при добавлении данных: {e}")

def get_products_by_country():
    """Функция для поиска товаров по стране с проверками"""
    try:
        country = input("Введите название страны: ").strip()
        if not country or country.isspace():
            print("Ошибка: название страны не может быть пустым")
            return

        data = read_data()

        if not data:
            print("Нет данных для поиска.")
            return

        found = False
        for c in data:
            if c.lower() == country.lower():
                country = c  # Сохраняем оригинальное написание
                found = True
                break

        if found:
            print(f"\n{country} экспортирует:")
            for product, quantity in sorted(data[country].items()):
                print(f"- {product}: {quantity} тонн")
        else:
            print("Страна не найдена.")

    except Exception as e:
        print(f"Произошла непредвиденная ошибка при поиске товаров по стране: {e}")

def get_countries_by_product():
    """Функция для поиска стран по товару с проверками"""
    try:
        product = input("Введите название товара: ").strip()
        if not product or product.isspace():
            print("Ошибка: название товара не может быть пустым")
            return

        data = read_data()

        if not data:
            print("Нет данных для поиска.")
            return

        countries = []
        for country, products in data.items():
            for p in products:
                if p.lower() == product.lower():
                    countries.append((country, products[p]))
                    break

        if countries:
            print(f"\n{product.capitalize()} экспортируют:")
            for country, quantity in sorted(countries):
                print(f"- {country}: {quantity} тонн")
        else:
            print("Товар не найден.")

    except Exception as e:
        print(f"Произошла непредвиденная ошибка при поиске стран по товару: {e}")

def delete_data():
    """Функция для удаления данных"""
    try:
        country = input("Введите страну для удаления (или оставьте пустым для удаления товара): ").strip()
        product = input("Введите товар для удаления (или оставьте пустым для удаления всей страны): ").strip()

        data = read_data()

        if not data:
            print("Нет данных для удаления.")
            return

        if country:
            if country in data:
                if product:
                    if product in data[country]:
                        del data[country][product]
                        print(f"Удален товар '{product}' из страны '{country}'")
                        if not data[country]:
                            del data[country]
                            print(f"Страна '{country}' удалена, так как больше не содержит товаров")
                    else:
                        print(f"Товар '{product}' не найден в стране '{country}'")
                else:
                    del data[country]
                    print(f"Удалена вся информация по стране '{country}'")
            else:
                print(f"Страна '{country}' не найдена")
        elif product:
            print("Не указана страна для удаления товара")
        else:
            print("Не указаны данные для удаления")

        write_data(data)

    except Exception as e:
        print(f"Произошла непредвиденная ошибка при удалении данных: {e}")

def menu():
    """Функция для отображения меню и обработки выбора пользователя"""
    while True:
        print("\nМеню:")
        print("1. Вывести всю информацию")
        print("2. Добавить данные")
        print("3. Найти товары по стране")
        print("4. Найти страны по товару")
        print("5. Удалить данные")
        print("6. Выход")

        choice = input("Выберите действие (1-6): ").strip()

        if choice == "1":
            show_all()
        elif choice == "2":
            add_data()
        elif choice == "3":
            get_products_by_country()
        elif choice == "4":
            get_countries_by_product()
        elif choice == "5":
            delete_data()
        elif choice == "6":
            print("Выход из программы.")
            break
        else:
            print("Неверный ввод! Пожалуйста, выберите пункт от 1 до 6.")

# Запуск меню
if __name__ == "__main__":
    menu()
```
