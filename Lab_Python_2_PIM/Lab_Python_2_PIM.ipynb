**1.** Пользователь вводит строку, символ (букву) и подстроку. В исходной строке после каждой заданной пользователем буквы вставьте заданную подстроку.

```python
input_string = input("Введите строку: ")
input_char = input("Введите символ: ")
input_substring = input("Введите подстроку: ")

# Инициализация результирующей строки
result = ""

# Цикл по строке
for i in range(len(input_string)):
    result += input_string[i]  # Добавляем текущий символ
    if input_string[i] == input_char:
        result += input_substring  # Если символ совпадает, добавляем подстроку

# Вывод результата
print("Измененная строка:", result)

import re

input_string = input("Введите строку: ")
input_char = input("Введите символ: ")
input_substring = input("Введите подстроку: ")

# Регулярное выражение для поиска всех вхождений символа
symbol = re.escape(input_char)

# Замена каждого вхождения символа на символ + подстрока
result = re.sub(symbol, input_char + input_substring, input_string)

# Вывод результата
print("Измененная строка:", result)
```

**2.** Подсчитайте сумму целых чисел, встречающихся в строке. Символ ‘–’ перед числом считается знаком отрицательного числа. Все остальные символы (кроме цифр и минуса) считаются разделителями между числами.

```python
input_string = input("Введите строку: ")

total_sum = 0
current_number = ''
is_negative = False

# Перебираем каждый символ в строке
for char in input_string:
    if char == '-':
        is_negative = True
    elif char.isdigit():
        current_number += char
    else:
        if current_number:
            if is_negative:
                total_sum -= int(current_number)
                is_negative = False
            else:
                total_sum += int(current_number)
            current_number = ''

# Обработка последнего числа (если оно есть)
if current_number:
    if is_negative:
        total_sum -= int(current_number)
    else:
        total_sum += int(current_number)

# Вывод результата
print("Сумма чисел в строке:", total_sum)

import re

input_string = input("Введите строку: ")

# Регулярное выражение для поиска всех целых чисел (положительных и отрицательных)
pattern = r'-?\d+'

# Найдем все числа в строке
numbers = re.findall(pattern, input_string)

# Преобразуем найденные строки в числа и подсчитаем сумму
total_sum = sum(int(num) for num in numbers)

# Вывод результата
print("Сумма чисел в строке:", total_sum)
```

**6.** Выведите слова сообщения, являющиеся палиндромами, по убыванию их длин. Для повторяющихся слов в скобках указывать количество повторов.

```python
message = "Мадам, я видел радар, а вы видели мадам?"
words = message.lower().split()

palindromes = {}
for word in words:
    word = ''.join(e for e in word if e.isalnum())
    if word == word[::-1]:
        if word in palindromes:
            palindromes[word] += 1
        else:
            palindromes[word] = 1

for word, count in sorted(palindromes.items(), key=lambda x: len(x[0]), reverse=True):
    if count > 1:
        print(f"{word} ({count})")
    else:
        print(word)


import re

message = "Мадам, я видел радар, а вы видели мадам?"
words = re.findall(r'\b\w+\b', message.lower())

palindromes = {}
for word in words:
    if word == word[::-1]:
        if word in palindromes:
            palindromes[word] += 1
        else:
            palindromes[word] = 1

for word, count in sorted(palindromes.items(), key=lambda x: len(x[0]), reverse=True):
    if count > 1:
        print(f"{word} ({count})")
    else:
        print(word)
```

```python
import re

string = input("Введите строку: ")

# Нормализация пробелов
string = re.sub(r'\s+', ' ', string).strip()

# Добавляем пробел перед открывающими скобками
string = re.sub(r'(\S)([({\[])', r'\1 \2', string)

# Удаляем пробелы внутри скобок
string = re.sub(r'([({[])\s+', r'\1', string)  # После открывающих
string = re.sub(r'\s+([)\]}])', r'\1', string)  # Перед закрывающими

# Удаляем пробелы перед знаками препинания
string = re.sub(r'\s+([.,!?;:])', r'\1', string)

# Добавляем пробел после знаков препинания и закрывающих скобок
string = re.sub(r'([.,!?;:…)\]}])(?=\S)', r'\1 ', string)

# Удаляем пробел между закрывающей скобкой и знаком препинания
string = re.sub(r'([)\]}]) ([.,!?;:])', r'\1\2', string)

# Удаляем пробел между закрывающей скобкой и другой скобкой
string = re.sub(r'([)\]}]) ([)\]}])', r'\1\2', string)

# Удаляем пробелы между несколькими скобками рядом
string = re.sub(r'([({[])\s+([({[])', r'\1\2', string)  # Между открывающими скобками
string = re.sub(r'([)\]}])\s+([)\]}])', r'\1\2', string)  # Между закрывающими скобками

# Исправляем многоточие
string = re.sub(r'\.\s+\.\s+\.', '...', string)

print("Исправленная строка:")
print(string)
#преар .оногб ,огенпог( г )   !прпрю    .    .  .ро( е ) .п ( ( ( (( р )) )  ) )   )   ?п(р)г  Ж             :па (  (  в )рю.
```

```python
def normalize_spaces(s):
    # Нормализация пробелов
    s = ' '.join(s.split())

    # Обработка скобок и пунктуации
    result = []
    i = 0
    n = len(s)
    while i < n:
        c = s[i]

        # Добавление пробела перед открывающей скобкой
        if c in '({[' and i > 0 and s[i-1] != ' ':
            result.append(' ')

        result.append(c)

        # Удаление пробелов внутри скобок
        if c in '({[':
            while i+1 < n and s[i+1] == ' ':
                i += 1
        elif c in ')}]':
            while len(result) > 1 and result[-2] == ' ':
                result.pop(-2)

        # Обработка пунктуации
        if c in '.,!?;:':
            # Удаление пробела перед пунктуацией
            while len(result) > 1 and result[-2] == ' ':
                result.pop(-2)
            # Добавление пробела после пунктуации
            if i+1 < n and s[i+1] != ' ':
                result.append(' ')

        i += 1

    # Объединение и исправление многоточия
    s = ''.join(result).replace('. . .', '...')
    return s

string = input("Введите строку: ")
print("Исправленная строка:")
print(normalize_spaces(string))
```
