```python
numbers=map(int,input().split())
numbers=list(numbers)
numbers_len = len(numbers)

sequences = []
tmp = []

i = 0
while i < numbers_len - 1:
    if numbers[i] > numbers[i + 1]:
        tmp.append(numbers[i])
    else:
        tmp.append(numbers[i])
        if len(tmp) > 1:
            sequences.append(tmp)
        tmp = []
    i += 1

tmp.append(numbers[i])
if len(tmp) > 1:
    sequences.append(tmp)

longest_sequence = max(sequences)

print("Самая длинная убывающая последовательность:", longest_sequence)
print("Её длина:", len(longest_sequence))
```

```python
# Ввод чисел
numbers = list(map(int, input("Введите числа через пробел: ").split()))
numbers_len = len(numbers)

sequences = []  # Список всех убывающих последовательностей
tmp = []        # Текущая убывающая последовательность

# Поиск всех убывающих последовательностей
for i in range(numbers_len):
    if not tmp:
        tmp.append(numbers[i])
    elif numbers[i] < tmp[-1]:
        tmp.append(numbers[i])
    else:
        if len(tmp) > 1:
            sequences.append(tmp)
        tmp = [numbers[i]]

# Проверка последней последовательности
if len(tmp) > 1:
    sequences.append(tmp)

# Поиск максимальной длины
if sequences:
    max_len = max(len(seq) for seq in sequences)
    longest_sequences = [seq for seq in sequences if len(seq) == max_len]

    print("Исходный список:", numbers)
    print("Длина самой длинной убывающей последовательности:", max_len)
    print("Самые длинные убывающие последовательности:")
    for seq in longest_sequences:
        print(seq)
else:
    print("Убывающих последовательностей не найдено.")
```

```python
numbers = list(map(int, input("Введите числа через пробел: ").split()))
A = int(input("Введите A (от 1 до 100): "))

delete_numbers = []
remaining_numbers = []

for number in numbers:
    if A / 6 < number < 5 * A / 6:
        delete_numbers.append(number)
    else:
        remaining_numbers.append(number)

print("Удалённые числа:", delete_numbers)
print("Оставшиеся числа:", remaining_numbers)
```

```python
def is_integer(s):
    return s.lstrip('-').isdigit()  # Проверяет, является ли строка целым числом (включая отрицательные)

def get_numbers():
    while True:
        input_str = input("Введите числа через пробел: ").strip()  # Удаляем лишние пробелы по краям
        if not input_str:  # Если строка пустая (пользователь ввёл "" или "   ")
            print("Ошибка! Введите хотя бы одно число.")
            continue  # Пропускаем оставшийся код и запрашиваем ввод снова

        parts = input_str.split()  # Разбиваем строку по пробелам
        if all(is_integer(num) for num in parts):  # Проверяем, все ли элементы — числа
            return list(map(int, parts))  # Если да, возвращаем список чисел
        else:
            print("Ошибка! Вводите только целые числа, разделённые пробелами.")

def get_A():
    while True:
        A_str = input("Введите A (от 1 до 100): ")
        if is_integer(A_str):
            A = int(A_str)
            if 1 <= A <= 100:
                return A
            else:
                print("Ошибка! A должно быть в диапазоне от 1 до 100.")
        else:
            print("Ошибка! Пожалуйста, введите целое число.")

# Получаем числа с проверкой
numbers = get_numbers()
A = get_A()

delete_numbers = []
remaining_numbers = []

for number in numbers:
    if A / 6 < number < 5 * A / 6:
        delete_numbers.append(number)
    else:
        remaining_numbers.append(number)

print("Удалённые числа:", delete_numbers)
print("Оставшиеся числа:", remaining_numbers)
```

**5.**	Напишите программу, на вход которой подаётся список чисел одной строкой. Программа должна для каждого элемента этого списка вывести сумму двух его соседей. Для элементов списка, являющихся крайними, одним из соседей считается элемент, находящий на противоположном конце этого списка. Например, если на вход подаётся список 1 3 5 6 10, то на выход ожидается список 13 6 9 15 7. Если на вход пришло только одно число, надо вывести его же. Вывод должен содержать одну строку с числами нового списка, разделёнными пробелом.

```python
numbers=map(int,input().split())
numbers=list(numbers)
# Проверка на случай, если список содержит только один элемент
if len(numbers) == 1:
    print(numbers[0])
else:
    # Создаем список для результата
    result = []

    # Проходим по каждому элементу списка
    for i in range(len(numbers)):
        # Определяем соседей с учетом циклического сдвига
        left_neighbor = numbers[i - 1]  # Левый сосед (индекс -1 для первого элемента)
        right_neighbor = numbers[(i + 1) % len(numbers)]  # Правый сосед (циклический индекс)

        # Сумма соседей
        result.append(left_neighbor + right_neighbor)

    # Вывод результата в виде строки
    print(' '.join(map(str, result)))
```

```python
def is_integer(s):
    """Проверяет, является ли строка целым числом (включая отрицательные)"""
    return s.lstrip('-').isdigit()

def get_numbers():
    """Запрашиваем список чисел с проверкой ввода"""
    while True:
        input_str = input("Введите целые числа через пробел: ").strip()
        if not input_str:
            print("Ошибка! Введите хотя бы одно число.")
            continue

        parts = input_str.split()
        if all(is_integer(num) for num in parts):
            return list(map(int, parts))
        else:
            print("Ошибка! Вводите только целые числа, разделённые пробелами.")

def main():
    # Ввод данных с проверкой
    numbers = get_numbers()

    # Находим индексы первого положительного и последнего отрицательного элементов
    first_positive_index = None
    last_negative_index = None

    for i, num in enumerate(numbers):
        if num > 0 and first_positive_index is None:
            first_positive_index = i
        if num < 0:
            last_negative_index = i

    # Проверяем, что индексы найдены
    if first_positive_index is not None and last_negative_index is not None and first_positive_index <= last_negative_index:
        # Сортируем элементы между первым положительным и последним отрицательным
        if first_positive_index < last_negative_index:
            numbers[first_positive_index + 1:last_negative_index] = sorted(numbers[first_positive_index + 1:last_negative_index])

    # Выводим результаты с проверкой на существование элементов
    if first_positive_index is not None:
        print("Первый положительный элемент:", numbers[first_positive_index], "с индексом", first_positive_index)
    else:
        print("В списке нет положительных элементов")

    if last_negative_index is not None:
        print("Последний отрицательный элемент:", numbers[last_negative_index], "с индексом", last_negative_index)
    else:
        print("В списке нет отрицательных элементов")

    print("Итоговый список:", numbers)

if __name__ == "__main__":
    main()
```

**9.**	Даны два неубывающих списка. Найдите их «пересечение», т. е. список, содержащий общие элементы исходных списков, причем кратность каждого элемента в итоговом списке равняется минимуму из его кратностей в исходных списках.

```python
numbers_1=list(map(int,input().split()))
numbers_2=list(map(int,input().split()))
crossing = []
i, j = 0, 0
while i < len(numbers_1) and j < len(numbers_2):
  if numbers_1[i] == numbers_2[j]:
    # Добавляем общий элемент
    crossing.append(numbers_1[i])
    i += 1
    j += 1
  elif numbers_1[i] < numbers_2[j]:
    i += 1
  else:
    j += 1
print(crossing)
```

```python
numbers = list(map(int, input("Введите неубывающий список положительных чисел: ").split()))
number = int(input("Введите нужное число: "))

left = 0
right = 0
current_sum = 0

while right <= len(numbers):
    if current_sum == number:
        print("Сумма элементов списка, составляющих данное число:", numbers[left:right])
        break
    elif current_sum < number:
        if right < len(numbers):
            current_sum += numbers[right]
            right += 1
        else:
            break  # Достигли конца списка, дальше нельзя увеличивать сумму
    else:
        current_sum -= numbers[left]
        left += 1
else:
    print("Сумма не найдена.")
```
