```python
num = int(input("Введите трёхзначное число: "))
if num < 100 or num > 999:
    print("Ошибка: это не трёхзначное число!")
    exit()

# Шаг 1: разбираем число на цифры
A = num // 100       # первая цифра
B = (num // 10) % 10 # вторая цифра
C = num % 10         # третья цифра

# Шаг 2: собираем все 6 перестановок и проверяем условия
"""6 перестановок, т.к. количество возможных перестановок
может 3! = 3*2*1, три разные цифры сотен, остаётся 2 разные
цифры десятков и наконец 1 последняя цифра единиц."""
# Перестановка 1: ABC
n1 = A * 100 + B * 10 + C
if A != 0:
    print(n1)

# Перестановка 2: ACB. Сравниваем с n1
n2 = A * 100 + C * 10 + B
if A != 0 and n2 != n1:  # проверка на трёхзначность и уникальность
    print(n2)

# Перестановка 3: BAC. Сравниваем с n1 и n2
n3 = B * 100 + A * 10 + C
if B != 0 and n3 != n1 and n3 != n2:
    print(n3)

# Перестановка 4: BCA. Сравниваем с n1, n2, n3
n4 = B * 100 + C * 10 + A
if B != 0 and n4 != n1 and n4 != n2 and n4 != n3:
    print(n4)

# Перестановка 5: CAB. Сравниваем с n1, n2, n3, n4
n5 = C * 100 + A * 10 + B
if C != 0 and n5 != n1 and n5 != n2 and n5 != n3 and n5 != n4:
    print(n5)

# Перестановка 6: CBA. Сравниваем с n1, n2, n3, n4, n5
n6 = C * 100 + B * 10 + A
if C != 0 and n6 != n1 and n6 != n2 and n6 != n3 and n6 != n4 and n6 != n5:
    print(n6)
```

**6.**	Найдите все пары дружественных чисел, не превышающих `А`. Два числа называют дружественными, если каждое из них равно сумме всех делителей другого (само число в качестве делителя не рассматривается). Числа в каждой паре должны идти по возрастанию.

```python
a=int(input ())
for i in range(1, a):
   k = 0
   n = 0
   for x in range(1, i):
       if i % x == 0:
           k += x
   for j in range(1, k):
       if k % j == 0:
           n += j
   if i == n and i != k and i == min(i, k) and i>=a and k>=a:
       print(i, k)
```

```python
a = int(input())  # Читаем значение A
found_pair = False  # Флаг для проверки наличия хотя бы одной пары
for i in range(1, a + 1):  # Итерируемся до A включительно
    k = 0
    for x in range(1, i):
        if i % x == 0:
            k += x
    if k > i and k <= a:  # Проверяем, что k больше i и не превышает a
        n = 0
        for j in range(1, k):
            if k % j == 0:
                n += j
        if i == n:
            print(i, k)
            found_pair = True  # Нашли пару
# Проверяем, были ли найдены подходящие пары
if not found_pair:
    print("Таких чисел нет")
```
