**1.** Сформировать одномерный массив, состоящий из n вещественных элементов co значением от -10 до 10.0. Для случайного числа y, такого, что *минимальный элемент массива < y < максимальный элемент массива*, вычислить:  
  
* произведение элементов массива, значения модуля которых больше модуля y,  
  
* сумму модулей остальных элементов.

```python
import numpy as np
import random

try:
    # Ввод числа n
    n = int(input("Введите количество элементов (целое число): "))
    if n <= 0:
        raise ValueError("Количество элементов должно быть положительным.")

    mult = 1
    total_sum = 0

    # Генерация массива случайных вещественных чисел
    a = [random.uniform(-10.0, 10.0) for _ in range(n)]
    a = np.around(a, decimals=1)

    min_a = np.min(a)
    max_a = np.max(a)

    # Генерируем y между min и max (исключая границы)
    if min_a < max_a:
        y = random.uniform(min_a, max_a)  # генерирует случайное вещественное число y в диапазоне min_a, max_a
    else:
        # Если все элементы одинаковые, генерируем y = 0 (или другое значение)
        y = 0.0

    # Проходим по каждому элементу массива
    for element in a:
        if abs(element) > abs(y):
            mult *= round(element, 1)
        else:
            total_sum += abs(round(element, 1))  # Суммируем модули

    # Вывод результатов
    print(a)
    print("Минимальное значение:", min_a)
    print("Случайное значение y:", y)
    print("Произведение элементов, модуль которых больше |y|:", round(mult, 1))
    print("Сумма модулей элементов, модуль которых меньше или равен |y|:", total_sum)

except ValueError as ve:
    print(f"Ошибка ввода: {ve}")
except Exception as e:
    print(f"Произошла непредвиденная ошибка: {e}")
```

```python
import numpy as np
import random

try:
    # Ввод числа n
    n = int(input("Введите количество элементов (целое число): "))
    if n <= 0:
        raise ValueError("Количество элементов должно быть положительным.")

    # Генерация массива случайных вещественных чисел и округление
    a = np.round(np.random.uniform(-10.0, 10.0, n), 1)

    min_a = np.min(a)
    max_a = np.max(a)

    # Генерация y между min и max (исключая границы)
    if min_a < max_a:
        y = random.uniform(min_a, max_a)
    else:
        y = 0.0

    # Векторизованные операции
    abs_a = np.abs(a)
    abs_y = abs(y)

    mask_gt = abs_a > abs_y
    mask_le = ~mask_gt

    mult = np.prod(a[mask_gt])
    total_sum = np.sum(abs_a[mask_le])

    # Вывод результатов
    print("Массив:", a)
    print("Минимальное значение:", min_a)
    print("Максимальное значение:", max_a)
    print("Случайное значение y:", round(y, 1))
    print("Произведение элементов, модуль которых больше |y|:", round(mult, 1))
    print("Сумма модулей элементов, модуль которых меньше или равен |y|:", round(total_sum, 1))

except ValueError as ve:
    print(f"Ошибка ввода: {ve}")
except Exception as e:
    print(f"Произошла непредвиденная ошибка: {e}")
```

```python
import numpy as np

def find_max_after_zero(arr):
    """Находит максимальный элемент после нуля"""
    try:
        # Преобразуем в numpy array, если это еще не массив
        arr = np.asarray(arr)

        # Проверка на пустой массив (работает для любого количества измерений)
        if arr.size == 0:
            raise ValueError("Массив пустой - нет элементов для анализа")

        if arr.ndim != 1:
            raise ValueError(f"Ожидается одномерный массив, получена размерность {arr.ndim}")

        # Находим индексы нулевых элементов (кроме последнего)
        zero_indices = np.where(arr[:-1] == 0)[0]

        # Получаем элементы, следующие за нулями
        elements_after_zero = arr[zero_indices + 1]

        # Проверка наличия элементов после нулей
        if len(elements_after_zero) == 0:
            raise ValueError("В массиве нет элементов, следующих за нулями")

        return np.max(elements_after_zero)

    except Exception as e:
        raise ValueError(f"Ошибка при обработке массива: {str(e)}")

# Тестовые случаи
test_cases = [
    np.array([5, 7, 0, 2, 3, 0, 5, 7, 0, 9, 7, 2]),  # Нормальный случай
    np.array([]),                                    # Пустой массив
    np.array([0, 0, 0]),                             # Нули в начале
    np.array([1, 2, 3]),                             # Нет нулей
    np.array([[1, 2], [3, 0], [3, 4]]),              # 2D массив
    np.array([0, 1, 0, 2, 0, 3]),                    # Несколько элементов после нулей
    [],                                              # Пустой список
    [0, 1, 0, 2, 0, 3]                               # Обычный список
]

for i, test_arr in enumerate(test_cases, 1):
    print(f"\nТест #{i}: {test_arr}")
    try:
        result = find_max_after_zero(test_arr)
        print(f"Максимальный элемент после нуля: {result}")
    except ValueError as e:
        print(f"{str(e)}")
```

```python
import numpy as np

def find_max_after_zero(arr):
    """Находит максимальный элемент после нуля"""
    try:
        arr = np.asarray(arr)

        if arr.ndim != 1:
            raise ValueError("Требуется одномерный массив")

        if arr.size == 0:
            raise ValueError("Массив пуст")

        zero_positions = np.where(arr == 0)[0]
        if zero_positions.size == 0:
            return "Нет нулей в массиве"

        valid_positions = zero_positions[zero_positions < len(arr)-1]
        if valid_positions.size == 0:
            return "Нет элементов после нулей"

        after_zero = arr[valid_positions + 1]
        return f"Максимальный элемент после нуля: {np.max(after_zero)}"

    except ValueError as e:
        if "could not convert" in str(e):
            raise ValueError("Массив должен содержать только числа")
        raise
    except Exception as e:
        raise ValueError(f"Ошибка обработки: {e}")

# Правильное тестирование
try:
    # Правильный синтаксис:
    test_arr = np.array([0])  # Запятые обязательны!
    print("Тестируем корректный массив:", test_arr)
    print(find_max_after_zero(test_arr))

except ValueError as e:
    print("Ошибка:", e)
```

```python
import numpy as np

def input_matrix():
    """Ввод матрицы"""
    print("Введите матрицу построчно (числа через пробел, пустая строка - конец):")
    rows = []
    while True:
        row = input().strip()
        if not row: break
        rows.append(list(map(int, row.split())))
    if not rows: raise ValueError("Матрица не может быть пустой")
    if len(set(len(row) for row in rows)) > 1: raise ValueError("Строки разной длины")
    return np.array(rows)

def analyze(matrix):
    """Анализ матрицы"""
    no_zero = np.sum(~np.any(matrix == 0, axis=1))
    vals, counts = np.unique(matrix, return_counts=True)
    max_rep = np.max(vals[counts > 1]) if np.any(counts > 1) else None
    return no_zero, max_rep

try:
    m = input_matrix()
    print(f"Строк без нулей: {analyze(m)[0]}\nМаксимальное повторное знаечение: {analyze(m)[1] or 'нет'}")
except Exception as e:
    print(f"Ошибка: {e}")
```

```python
import numpy as np
from numpy.lib.stride_tricks import sliding_window_view

def load_and_validate_matrix(filepath):
    """Загрузка и валидация матрицы"""
    try:
        data = np.loadtxt(filepath, ndmin=2)
        if data.shape != (10, 10):
            raise ValueError(f"Ожидается матрица 10x10, получена {data.shape[0]}x{data.shape[1]}")
        return data.astype(float)
    except ValueError as e:
        if "could not convert string to float" in str(e):
            raise ValueError("Обнаружены нечисловые значения в данных")
        raise ValueError(f"Ошибка формата данных")
    except FileNotFoundError:
        raise ValueError("Файл не найден")
    except Exception as e:
        raise ValueError(f"Ошибка при чтении файла: {str(e)}")

def count_local_minima(matrix):
    """Подсчёт локальных минимумов (в широком смысле) без циклов"""
    try:
        # Добавляем границу из бесконечностей
        padded = np.pad(matrix, 1, mode='constant', constant_values=np.inf)
        # Получаем все окрестности 3x3
        neighborhoods = sliding_window_view(padded, (3, 3)).reshape(10, 10, 9)
        # Убираем центральный элемент из каждого окна
        neighbors = np.delete(neighborhoods, 4, axis=2)
        # Сравниваем центральные элементы со всеми соседями
        return np.sum(np.all(matrix[:, :, np.newaxis] < neighbors, axis=2))
    except Exception as e:
        raise ValueError(f"Ошибка при поиске минимумов: {str(e)}")

def main():
    filepath = '/content/drive/MyDrive/Программирование на языке Python/Lab_Python_5_PIM/input.txt'
    try:
        print("Загрузка матрицы...")
        matrix = load_and_validate_matrix(filepath)
        print("\nМатрица успешно загружена:")
        [print(" ".join(f"{x:4.0f}" for x in row)) for row in matrix]
        print("\nПодсчёт локальных минимумов...")
        minima = count_local_minima(matrix)
        print(f"\nКоличество локальных минимумов: {minima}")
    except ValueError as e:
        print(f"\n{str(e)}")
        print("\nТребования к файлу:")
        print("- Ровно 10 строк, по 10 чисел в каждой")
        print("- Числа разделены пробелами")

if __name__ == "__main__":
    main()
```

**7.** Удалите строку и столбец, на пересечении которых находится минимальный элемент массива. После этого отсортируйте столбцы массива по неубыванию минимальных элементов в них.

```python
import numpy as np

try:
    # Создаем простой массив (можно изменить)
    matrix = np.array([[9, 5, 2],
                       [7, 3, 8],
                       [4, 2, 1]])

    print("Исходный массив:\n", matrix)

    # Проверка на пустой массив
    if matrix.size == 0:
        print("Ошибка: массив пустой.")
        exit()

    # 1. Находим минимальный элемент и его координаты
    try:
        min_val = np.min(matrix)
        min_row, min_col = np.where(matrix == min_val)
        min_row = min_row[0]
        min_col = min_col[0]
        print(f"Минимальный элемент: {min_val}, строка: {min_row+1}, столбец: {min_col+1}")
    except IndexError:
        print("Ошибка при определении координат минимального элемента.")
        exit()
    except Exception as e:
        print(f"Ошибка при поиске минимума: {e}")
        exit()

    # 2. Удаляем строку и столбец
    try:
        matrix = np.delete(matrix, min_row, axis=0)  # Удаление строки
        matrix = np.delete(matrix, min_col, axis=1)  # Удаление столбца
        print("Массив после удаления строки и столбца:\n", matrix)
    except IndexError:
        print("Ошибка при удалении строки или столбца — возможно, неверные индексы.")
        exit()
    except Exception as e:
        print(f"Ошибка при удалении строки/столбца: {e}")
        exit()

    # 3. Сортировка столбцов по минимальным элементам
    try:
        if matrix.size == 0 or matrix.shape[1] == 0:
            print("Невозможно отсортировать — массив пуст после удаления.")
            exit()

        min_in_cols = np.min(matrix, axis=0)
        sorted_indices = np.argsort(min_in_cols)
        matrix = matrix[:, sorted_indices]
        print("Отсортированный массив (по минимальным элементам столбцов):\n", matrix)
    except Exception as e:
        print(f"Ошибка при сортировке столбцов: {e}")
        exit()

except Exception as e:
    print(f"Непредвиденная ошибка: {e}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import re
from collections import Counter
import os
import numpy as np

# Путь к папке с текстами
texts_folder = '/content/drive/MyDrive/Программирование на языке Python/Lab_Python_5_PIM/texts/'

# Попытка получить список файлов
try:
    files = os.listdir(texts_folder)
    if not files:
        print("Папка пуста.")
        exit()
except FileNotFoundError:
    print(f"Папка '{texts_folder}' не найдена.")
    exit()
except PermissionError:
    print("Нет доступа к папке.")
    exit()
except Exception as e:
    print(f"Ошибка при получении списка файлов: {e}")
    exit()

def plot_word_frequency(file_number):
    try:
        # Поиск файла по номеру
        file_name = next((f for f in files if f.startswith(file_number)), None)

        if not file_name:
            print(f"Файл с номером '{file_number}' не найден.")
            return

        file_path = os.path.join(texts_folder, file_name)

        # Чтение и обработка текста
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read().lower()

        words = re.findall(r'\b\w{3,}\b', text)  # Игнорируем слова короче 3 букв
        if not words:
            print("В файле нет подходящих слов для анализа.")
            return

        word_counts = Counter(words)
        top_50 = word_counts.most_common(50)

        # Подготовка данных для графика
        words, counts = zip(*top_50)
        x = np.arange(len(words))

        # Настройка стиля (используем доступный стиль)
        plt.style.use('seaborn-v0_8')  # Или другой доступный стиль из plt.style.available

        # Создание фигуры с увеличенным размером
        fig, ax = plt.subplots(figsize=(22, 12))

        # Столбчатая диаграмма с цветами по частоте
        colors = plt.cm.plasma(np.linspace(0, 1, len(counts)))
        bars = ax.bar(x, counts, color=colors)

        # Настройка осей и подписей
        ax.set_xticks(x)
        ax.set_xticklabels(words, rotation=75, ha='right', fontsize=10)
        ax.set_xlabel('Слова', fontsize=14)
        ax.set_ylabel('Частота', fontsize=14)

        # Добавление значений на столбцы (с умным размещением)
        max_count = max(counts)
        for i, (word, count) in enumerate(top_50):
            va = 'bottom' if count < max_count * 0.8 else 'top'
            ax.text(i, count, str(count),
                   ha='center', va=va, fontsize=9, rotation=45)

        # Настройка сетки и границ
        ax.yaxis.grid(True, linestyle='--', alpha=0.7)
        ax.set_axisbelow(True)
        plt.margins(y=0.1)

        # Заголовок с именем файла
        clean_title = ' '.join(re.findall(r'[a-zA-Zа-яА-Я]+', file_name.split('.')[0][len(file_number):]))
        ax.set_title(f'Топ 50 слов в тексте "{clean_title}"', pad=20, fontsize=16)

        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"Ошибка: {str(e)}")

# Вывод списка файлов и обработка ввода
print("Доступные файлы:")
for file in sorted(files, key=lambda x: int(re.match(r'\d+', x).group())):
    print(file)

try:
    file_num = input("\nВведите номер файла: ").strip()
    plot_word_frequency(file_num)
except Exception as e:
    print(f"Ошибка ввода: {str(e)}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/drive/MyDrive/Программирование на языке Python/Lab_Python_5_PIM/osadki.txt'

def validate_data(months, precipitation):
    """Проверка корректности данных перед построением графиков"""
    errors = []

    # Проверка количества месяцев
    if len(months) != 12:
        errors.append(f"Ожидается 12 месяцев, получено {len(months)}")

    # Проверка наличия всех месяцев
    required_months = ['Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь',
                      'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь']
    for month in required_months:
        if month not in months:
            errors.append(f"Отсутствует месяц: {month}")

    # Проверка значений осадков
    for i, amount in enumerate(precipitation):
        if not isinstance(amount, (int, float)):
            errors.append(f"Некорректное значение осадков для {months[i]}: {amount}")
        elif amount < 0:
            errors.append(f"Отрицательное значение осадков для {months[i]}: {amount}")
        elif amount > 1000:  # Реалистичная проверка верхнего предела
            errors.append(f"Нереалистично высокое значение осадков для {months[i]}: {amount}")

    return errors

try:
    # Считываем данные из файла
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            data = file.readlines()
    except FileNotFoundError:
        print(f"Ошибка: файл '{file_path}' не найден.")
        exit()
    except PermissionError:
        print(f"Ошибка: нет доступа к файлу '{file_path}'.")
        exit()
    except Exception as e:
        print(f"Ошибка при открытии файла: {e}")
        exit()

    # Обработка данных
    months = []
    precipitation = []

    for line_num, line in enumerate(data, start=1):
        try:
            if not line.strip():
                continue  # Пропускаем пустые строки
            parts = line.split()
            if len(parts) != 2:
                raise ValueError(f"Неверный формат в строке {line_num}: '{line.strip()}'")
            month, amount = parts
            months.append(month)
            precipitation.append(int(amount))
        except ValueError as ve:
            print(f"Ошибка в строке {line_num}: {ve}")
        except Exception as e:
            print(f"Непредвиденная ошибка в строке {line_num}: {e}")

    if not months or not precipitation:
        print("Ошибка: нет корректных данных для построения графика.")
        exit()

    # Проверка данных перед построением графиков
    validation_errors = validate_data(months, precipitation)
    if validation_errors:
        print("\nОбнаружены ошибки в данных:")
        for error in validation_errors:
            print(f"- {error}")
        print("\nРекомендации:")
        print("- Убедитесь, что файл содержит данные для всех 12 месяцев")
        print("- Названия месяцев должны быть полными (Январь, Февраль и т.д.)")
        print("- Значения осадков должны быть положительными числами")
        print("- Проверьте, что значения осадков выглядят реалистично")
        print("\nПример корректной строки: 'Январь 45'")
        exit()

    # Создание DataFrame
    df = pd.DataFrame({
        'Месяцы': months,
        'Осадки': precipitation
    })

    # Построение графиков (остальной код без изменений)
    try:
        fig, axs = plt.subplots(3, 1, figsize=(10, 15))

        # Столбчатая диаграмма
        axs[0].bar(df['Месяцы'], df['Осадки'], color='blue')
        axs[0].set_title('Годовое количество осадков (столбчатая диаграмма)')
        axs[0].set_ylabel('Осадки (мм)')
        axs[0].set_xlabel('Месяцы')

        # Горизонтальная диаграмма
        axs[1].barh(df['Месяцы'], df['Осадки'], color='green')
        axs[1].set_title('Годовое количество осадков (горизонтальная диаграмма)')
        axs[1].set_xlabel('Осадки (мм)')
        axs[1].set_ylabel('Месяцы')

        # Круговая диаграмма
        axs[2].pie(df['Осадки'], labels=df['Месяцы'], autopct='%1.1f%%', startangle=90)
        axs[2].set_title('Годовое количество осадков (круговая диаграмма)')

        plt.tight_layout()

        # Сохранение графика
        try:
            output_path = '/content/drive/MyDrive/Программирование на языке Python/Lab_Python_5_PIM/osadki.png'
            plt.savefig(output_path)
            print(f"График успешно сохранён в файл: {output_path}")
        except Exception as e:
            print(f"Ошибка при сохранении графика: {e}")

        # Показ графика
        plt.show()

    except Exception as e:
        print(f"Ошибка при построении графиков: {e}")

except KeyboardInterrupt:
    print("\nВыполнение прервано пользователем.")
except Exception as e:
    print(f"Непредвиденная ошибка: {e}")
```

**11.** Нарисовать звёздочками зелёного цвета график кардиоиды в полярной системе координат. Добавьте легенду с заголовком «Функция» и подписью «Кардиоида» в левый верхний угол.
Угол принадлежит промежутку от 0 до 3π/2, r = 2,5. Уравнение кардиоиды: ρ = 2*r*(1 + cosφ).



```python
# Импортируем библиотеки
import numpy as np
import matplotlib.pyplot as plt

try:
    # Задаем диапазон углов φ от 0 до 3π/2
    try:
        phi = np.linspace(0, 3 * np.pi / 2, 100)
    except Exception as e:
        print(f"Ошибка при создании массива углов φ: {e}")
        exit()

    # Уравнение кардиоиды: ρ = 2r(1 + cosφ)
    try:
        r = 2 * (1 + np.cos(phi))
    except Exception as e:
        print(f"Ошибка при вычислении значений r: {e}")
        exit()

    # Построение графика
    try:
        plt.figure(figsize=(8, 8))
        plt.polar(phi, r, color='green', linestyle='None', marker='*', markersize=10)  # Звёздочки

        plt.legend(['Кардиоида'], title='Функция', loc='upper left')
        plt.title('График кардиоиды')
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Ошибка при построении графика: {e}")
        exit()

except KeyboardInterrupt:
    print("\nВыполнение прервано пользователем.")
except Exception as e:
    print(f"Непредвиденная ошибка: {e}")
```

**12.** Написать программу изображения графика функции y=√x. Не забывайте про то, что функция имеет значения ≥ 0. Предварительно программа запрашивает у пользователя следующую информацию:

* цвет графика (программа предлагает не менее пяти различных цветов);
* тип линии (программа предлагает не менее трех различных вариантов);
* толщину линии (программа предлагает не менее пяти различных вариантов).

Все запросы к пользователю обрабатываются с использованием обработки исключительных ситуаций. Варианты можно вводить строкой или числом (на ваш выбор).


```python
import numpy as np
import matplotlib.pyplot as plt

def get_color():
    colors = ['red', 'green', 'blue', 'orange', 'purple']
    while True:
        print("Доступные цвета: red, green, blue, orange, purple")
        color = input("Введите цвет графика: ").strip().lower()
        if color in colors:
            return color
        else:
            print("Некорректный ввод. Попробуйте еще раз.")

def get_line_style():
    line_styles = ['-', '--', '-.', ':']
    while True:
        print("Доступные типы линий: '-', '--', '-.', ':'")
        line_style = input("Введите тип линии: ").strip()
        if line_style in line_styles:
            return line_style
        else:
            print("Некорректный ввод. Попробуйте еще раз.")

def get_line_width():
    while True:
        print("Доступные толщины линии: 1, 2, 3, 4, 5")
        try:
            line_width = int(input("Введите толщину линии (число от 1 до 5): "))
            if line_width in range(1, 6):
                return line_width
            else:
                print("Толщина линии должна быть от 1 до 5. Попробуйте еще раз.")
        except ValueError:
            print("Некорректный ввод. Пожалуйста, введите целое число.")

def plot_graph(color, line_style, line_width):
    try:
        x = np.linspace(0, 10, 100)
        y = np.sqrt(x)

        plt.plot(x, y, color=color, linestyle=line_style, linewidth=line_width)
        plt.title('График функции y = √x')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.xlim(0, 10)
        plt.ylim(0, np.sqrt(10))
        plt.grid()
        plt.show()

    except Exception as e:
        print(f"Произошла ошибка при построении графика: {str(e)}")

def main():
    try:
        print("Программа для построения графика функции y = √x")
        color = get_color()
        line_style = get_line_style()
        line_width = get_line_width()
        plot_graph(color, line_style, line_width)

    except KeyboardInterrupt:
        print("\nПрограмму прервали вручную.")

    except Exception as e:
        print(f"Произошла непредвиденная ошибка: {str(e)}")

if __name__ == "__main__":
    main()
```
