**1.** Разработайте программу на языке программирования Python, описывающую музыкальные записи в домашней коллекции. Любая запись характеризуется названием песни, исполнителем, названием альбома, годом выхода, длительностью. Предусмотреть возможность добавления записи в коллекцию, удаления записи, изменения данных о записи, вывода всех записей из указанного альбома определенного исполнителя, поиска записи по названию или длительности.

```python
class MusicRecord:
    def __init__(self, title, artist, album, year, duration):
        # Проверяем, что все параметры имеют правильный тип
        if not isinstance(title, str):
            raise TypeError("Название песни должно быть текстом")
        if not isinstance(artist, str):
            raise TypeError("Имя певца песни должно быть текстом")
        if not isinstance(album, str):
            raise TypeError("Название альбома должно быть текстом")
        if not isinstance(year, int):
            raise TypeError("Год должен быть числом")
        if not isinstance(duration, str):
            raise TypeError("Продолжительность должна быть быть текстом")

        self.title = title
        self.artist = artist
        self.album = album
        self.year = year
        self.duration = duration


class MusicCollection:
    def __init__(self):
        self.records = []

    def add_record(self, record):
        # Проверим, является ли объект экземпляром класса MusicRecord
        if not isinstance(record, MusicRecord):
            raise TypeError("Объект не является типом кламма MusicRecord")
        self.records.append(record)

    def remove_record(self, title):
        found = False
        for record in self.records:
            if record.title == title:
                self.records.remove(record)
                found = True
                break
        if not found:
            raise ValueError(f"Нет записи с названием '{title}' для удаления")

    def update_record(self, title, new_record):
        if not isinstance(new_record, MusicRecord):
            raise TypeError("Тип новой записи должен быть MusicRecord")

        found = False
        for i, record in enumerate(self.records):
            if record.title == title:
                self.records[i] = new_record
                found = True
                break
        if not found:
            raise ValueError(f"Нет записей с названием '{title}' для обновления")

    def get_records_by_album(self, artist, album):
        return [record for record in self.records if record.artist == artist and record.album == album]

    def search_record_by_title(self, title):
        results = [record for record in self.records if record.title == title]
        if len(results) == 0:
            raise ValueError(f"Не найдено ни одной записи с названием '{title}'")
        return results

    def search_record_by_duration(self, duration):
        results = [record for record in self.records if record.duration == duration]
        if len(results) == 0:
            raise ValueError(f"Не найдено ни одной записи с продолжительностью'{duration}'")
        return results


# Пример использования
if __name__ == "__main__":
    try:
        collection = MusicCollection()

        record1 = MusicRecord("Песня 1", "Артист 1", "Альбом 1", 2020, "3:30")
        record2 = MusicRecord("Песня 2", "Артист 1", "Альбом 1", 2020, "4:00")

        collection.add_record(record1)
        collection.add_record(record2)

        print("Все песни в Альбом 1 Артист 1:")
        for record in collection.get_records_by_album("Артист 1", "Альбом 1"):
            print(f"{record.title} у {record.artist}")

        print("\nПоиск 'Песня 1':")
        for record in collection.search_record_by_title("Песня 1"):
            print(f"{record.title} у {record.artist}")

        print("\nПоиск продолжительности '4:00':")
        for record in collection.search_record_by_duration("4:00"):
            print(f"{record.title} у {record.artist}")

        collection.remove_record("Песня 2")
        print("\nПосле удаления 'Песня 2':")
        for record in collection.records:
            print(f"{record.title} у {record.artist}")

    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
class Property:
    def __init__(self, address, size, rooms, rent):
        # Проверка адреса
        if not isinstance(address, str) or len(address.strip()) == 0:
            raise ValueError("Адрес должен быть непустой строкой")

        # Проверка площади
        if not isinstance(size, (int, float)) or size <= 0:
            raise ValueError("Площадь должна быть положительным числом")

        # Проверка количества комнат
        if not isinstance(rooms, int) or rooms <= 0:
            raise ValueError("Количество комнат должно быть целым положительным числом")

        # Проверка арендной платы
        if not isinstance(rent, (int, float)) or rent <= 0:
            raise ValueError("Арендная плата должна быть положительным числом")

        self.address = address.strip()
        self.size = size
        self.rooms = rooms
        self.rent = rent

    def display_info(self):
        print(f"Адрес: {self.address}")
        print(f"Площадь: {self.size} кв. м")
        print(f"Количество комнат: {self.rooms}")
        print(f"Арендная плата: {self.rent} рублей/месяц")


class RentalAgency:
    def __init__(self):
        self.properties = []

    def add_property(self, property):
        if not isinstance(property, Property):
            raise ValueError("Можно добавлять только объекты класса Property")
        self.properties.append(property)

    def find_property_by_criteria(self, min_size, min_rooms, max_rent):
        # Проверка критериев поиска
        if not isinstance(min_size, (int, float)) or min_size < 0:
            raise ValueError("Минимальная площадь должна быть неотрицательным числом")
        if not isinstance(min_rooms, int) or min_rooms < 0:
            raise ValueError("Минимальное количество комнат должно быть неотрицательным целым числом")
        if not isinstance(max_rent, (int, float)) or max_rent <= 0:
            raise ValueError("Максимальная арендная плата должна быть положительным числом")

        filtered_properties = [prop for prop in self.properties
                             if prop.size >= min_size
                             and prop.rooms >= min_rooms
                             and prop.rent <= max_rent]
        return filtered_properties


# Пример использования с обработкой ошибок
try:
    property1 = Property("г. Киров, ул. Ленина, д. 36, кв. 7", 23, 1, 20000)
    property2 = Property("г. Киров, ул. Воровского, д. 103, кв. 92", 57, 2, 45000)
    property3 = Property("г. Киров, ул. Московская, д. 15, кв. 34", 45.5, 3, 38000)

    agency = RentalAgency()
    agency.add_property(property1)
    agency.add_property(property2)
    agency.add_property(property3)

    print("Свободное жильё согласно критериям:")
    filtered_properties = agency.find_property_by_criteria(20, 1, 50000)

    if not filtered_properties:
        print("По заданным критериям ничего не найдено.")
    else:
        for prop in filtered_properties:
            prop.display_info()
            print()  # Пустая строка между объектами

except ValueError as e:
    print(f"Ошибка в данных: {e}")
except Exception as e:
    print(f"Неожиданная ошибка: {e}")
```

```python
from typing import List, Dict, Optional, Union

class Property:
    def __init__(self, address: str, size: float, rooms: int, rent: float,
                 district: Optional[str] = None, has_furniture: bool = False,
                 floor: Optional[int] = None, max_floor: Optional[int] = None):
        self.address = address
        self.size = size
        self.rooms = rooms
        self.rent = rent
        self.district = district
        self.has_furniture = has_furniture
        self.floor = floor
        self.max_floor = max_floor

    def display_info(self) -> None:
        print(f"\nАдрес: {self.address}")
        print(f"Район: {self.district if self.district else 'не указан'}")
        print(f"Площадь: {self.size} кв. м")
        print(f"Количество комнат: {self.rooms}")
        print(f"Арендная плата: {self.rent} рублей/месяц")
        print(f"Мебель: {'есть' if self.has_furniture else 'нет'}")
        if self.floor and self.max_floor:
            print(f"Этаж: {self.floor}/{self.max_floor}")


class PropertyFilter:
    def __init__(self, **kwargs):
        self.criteria = kwargs

    def matches(self, property: Property) -> bool:
        for key, value in self.criteria.items():
            if not hasattr(property, key):
                return False
            prop_value = getattr(property, key)

            if prop_value is None and value is not None:
                return False

            if isinstance(value, (tuple, list)) and len(value) == 2:
                min_val, max_val = value
                if min_val is not None and prop_value < min_val:
                    return False
                if max_val is not None and prop_value > max_val:
                    return False
            elif isinstance(value, dict) and 'op' in value:
                op = value['op']
                if op == 'in' and prop_value not in value['values']:
                    return False
                elif op == 'contains' and value['substr'].lower() not in str(prop_value).lower():
                    return False
                elif op == '<' and prop_value >= value['value']:
                    return False
                elif op == '>' and prop_value <= value['value']:
                    return False
            elif prop_value != value:
                return False
        return True


class RentalAgency:
    def __init__(self):
        self.properties: List[Property] = []

    def add_property(self, property: Property) -> None:
        self.properties.append(property)

    def find_properties(self, property_filter: PropertyFilter) -> List[Property]:
        return [prop for prop in self.properties if property_filter.matches(prop)]

    def get_districts(self) -> List[str]:
        districts = set(prop.district for prop in self.properties if prop.district)
        return sorted(districts)

    def get_available_rooms(self) -> List[int]:
        rooms = set(prop.rooms for prop in self.properties)
        return sorted(rooms)


def init_agency() -> RentalAgency:
    agency = RentalAgency()

    agency.add_property(Property(
        "г. Киров, ул. Ленина, д. 36, кв. 7",
        23, 1, 20000,
        district="Центр",
        has_furniture=True,
        floor=3,
        max_floor=5
    ))

    agency.add_property(Property(
        "г. Киров, ул. Воровского, д. 103, кв. 92",
        57, 2, 45000,
        district="Заречный",
        has_furniture=False,
        floor=9,
        max_floor=9
    ))

    agency.add_property(Property(
        "г. Киров, ул. Московская, д. 15, кв. 34",
        45, 3, 35000,
        district="Центр",
        has_furniture=True,
        floor=2,
        max_floor=5
    ))

    agency.add_property(Property(
        "г. Киров, ул. Преображенская, д. 28, кв. 12",
        32, 1, 22000,
        district="Центр",
        has_furniture=False,
        floor=1,
        max_floor=10
    ))

    return agency


def get_user_filter(agency: RentalAgency) -> PropertyFilter:
    filters = {}

    print("\nВведите критерии поиска (оставьте поле пустым, если параметр не важен):")

    # Район
    districts = agency.get_districts()
    if districts:
        print(f"\nДоступные районы: {', '.join(districts)}")
        district = input("Район: ").strip()
        if district and district in districts:
            filters['district'] = district

    # Количество комнат
    rooms = agency.get_available_rooms()
    print(f"\nДоступные варианты по комнатам: {', '.join(map(str, rooms))}")
    room_input = input("Количество комнат (можно несколько через запятую): ").strip()
    if room_input:
        try:
            rooms_list = [int(r.strip()) for r in room_input.split(',')]
            filters['rooms'] = {'op': 'in', 'values': rooms_list}
        except ValueError:
            pass

    # Цена
    max_price = input("\nМаксимальная цена аренды: ").strip()
    if max_price:
        try:
            filters['rent'] = {'op': '<', 'value': float(max_price)}
        except ValueError:
            pass

    # Площадь
    min_size = input("\nМинимальная площадь (кв.м): ").strip()
    max_size = input("Максимальная площадь (кв.м): ").strip()
    if min_size or max_size:
        try:
            min_val = float(min_size) if min_size else None
            max_val = float(max_size) if max_size else None
            filters['size'] = (min_val, max_val)
        except ValueError:
            pass

    # Мебель
    furniture = input("\nНужна меблированная квартира? (да/нет): ").strip().lower()
    if furniture in ['да', 'yes']:
        filters['has_furniture'] = True
    elif furniture in ['нет', 'no']:
        filters['has_furniture'] = False

    # Этаж
    min_floor = input("\nМинимальный этаж: ").strip()
    max_floor = input("Максимальный этаж: ").strip()
    if min_floor or max_floor:
        try:
            min_val = int(min_floor) if min_floor else None
            max_val = int(max_floor) if max_floor else None
            filters['floor'] = (min_val, max_val)
        except ValueError:
            pass

    return PropertyFilter(**filters)


def main():
    agency = init_agency()

    while True:
        print("\n" + "="*50)
        print("Система поиска аренды жилья")
        print("1. Поиск недвижимости")
        print("2. Выход")

        choice = input("\nВыберите действие: ").strip()

        if choice == "1":
            prop_filter = get_user_filter(agency)
            results = agency.find_properties(prop_filter)

            if not results:
                print("\nПо вашим критериям ничего не найдено.")
                continue

            print(f"\nНайдено объектов: {len(results)}")
            for i, prop in enumerate(results, 1):
                print(f"\nРезультат #{i}:")
                prop.display_info()

            input("\nНажмите Enter чтобы продолжить...")

        elif choice == "2":
            print("До свидания!")
            break

        else:
            print("Некорректный ввод. Попробуйте снова.")


if __name__ == "__main__":
    main()
```
